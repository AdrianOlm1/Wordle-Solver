```python
## automatic solver to find average
import random
from collections import Counter

def solve(hiddenWord, rankedWordMod):
    ## removes all words without letter at specified index
    def removeCorrectLetter(letter, index):
        nonlocal rankedWordMod
        rankedWordMod = [entry for entry in rankedWordMod if entry[index] == letter]

    ## removes all words containing letter at index
    def removeAlmostLetter(letter, index):
        nonlocal rankedWordMod
        rankedWordMod = [entry for entry in rankedWordMod if letter in entry and entry[index] != letter]

    ## removes all words containing a letter
    def removeIncorrectLetter(letter):
        nonlocal rankedWordMod
        rankedWordMod = [entry for entry in rankedWordMod if letter not in entry]

    ## to handle similar words and not get stuck
    def greaterTwoCase(word):
        usedLetters = set(finalWord) - {'_'}  # skip confirmed letters

        # Build a pool of remaining letters not in the known correct positions
        remainingLetters = set()
        for word in rankedWordMod:
            for ch in word:
                if ch not in usedLetters:
                    remainingLetters.add(ch)
    
        # First, try to find a word with 5 unique letters from remaining letters
        for word in rankedWordList:
            if len(set(word)) == 5 and set(word).issubset(remainingLetters):
                return word
                
        # attempt to look for word with set of 4 letters
        for word in rankedWordList:
            # Filter the word to only contain letters in remainingLetters
            filtered_letters = set(c for c in word if c in remainingLetters)
            
            # Check if there are at least 4 unique letters in the word that are in remainingLetters
            if len(filtered_letters) >= 4:
                return word

        # attempt to look for word with set of 3 letters
        for word in rankedWordList:
            # Filter the word to only contain letters in remainingLetters
            filtered_letters = set(c for c in word if c in remainingLetters)
            
            # Check if there are at least 3 unique letters in the word that are in remainingLetters
            if len(filtered_letters) >= 3:
                return word
    
        # If all else fails
        return "nope"


    ## used when all letter guesses are in the word (indefinite)
    def lowerBoundRemove(letter, count):
        nonlocal rankedWordMod
        rankedWordMod = [
            word for word in rankedWordMod
            if word.count(letter) >= count
        ]

    ## used when we know x's amount of letter must be word (definite)
    def upperBoundRemove(letter, count):
        nonlocal rankedWordMod
        rankedWordMod = [
            word for word in rankedWordMod
            if word.count(letter) == count
        ]

    #rankedWordMod = rankedWordList.copy()
    correctCount = 0
    tried = False
    finalWord = ['_'] * 5

    for i in range(6):
        if i == 5:
            tried = True
        twoBranch = False

        if len(rankedWordMod) == 1:
            return i+1

        ## attempts to solve edge case of similar words
        if correctCount >= 3 and len(rankedWordMod) > 3 and not tried:
            bestWord = greaterTwoCase(finalWord)
            ## unable to find word using only unused words
            if bestWord != "nope":
                word = bestWord
                twoBranch = True
            ## found word using only unused words
            else:
                word = rankedWordMod[0]
                tried = True
                rankedWordMod.remove(word)
        ## grabs new word from top of rank as usual
        else:
            word = rankedWordMod[0]
            rankedWordMod.remove(word)

        guessCounts = Counter(word)
        actualCounts = Counter(hiddenWord)

        for j in range(5):
            ## gives our hidden and guess word
            hiddenLetter = hiddenWord[j]
            guessedLetter = word[j]

            ## to make sure we don't reask for known spots
            if finalWord[j] != '_' and not twoBranch:
                continue

            ## if the letter was correct
            if guessedLetter in hiddenWord:
                ## if the letter is in the correct spot
                if guessedLetter == hiddenLetter:
                    ## counts how many are in the correct spot to return early
                    if finalWord[j] != guessedLetter:
                        finalWord[j] = guessedLetter
                        correctCount += 1
                    if correctCount == 5:
                        return i+1
                    ## adds to our final word and removed a correct from possibilities
                    removeCorrectLetter(guessedLetter, j)
                else:
                    removeAlmostLetter(guessedLetter, j)

                ## to check if there are multiple all correct letters
                if guessCounts[guessedLetter] <= actualCounts[guessedLetter]:
                    lowerBoundRemove(guessedLetter, guessCounts[guessedLetter])
                ## to check if there are multiple correct letters with some incorrect
                elif guessCounts[guessedLetter] > actualCounts[guessedLetter]:
                    upperBoundRemove(guessedLetter, actualCounts[guessedLetter])

            ## if letter is not in the word at all
            else:
                removeIncorrectLetter(guessedLetter)
    return 7  # max attempts used


# Run the solver
sum = 0
fails = 0
size = 100000
failedWords = []
for i in range(size):
    random_word = random.choice(rankedWordList)
    num = solve(random_word, rankedWordList.copy())
    if num == 7:
        fails += 1
        failedWords.append(random_word)
    else: 
        sum += num
print(f"Solve Average: {sum/(size - fails)}, Fail Total: {fails}, , Fail Rate: {fails / size:.2%}") 
sort = list(set(failedWords))
print(sorted(sort))
```
